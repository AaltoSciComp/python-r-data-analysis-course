# Calculation of $\pi$ using a monte carlo method

In this project, I will calculate $\pi$ by a simple monte carlo method.  We will place random points in a square, and count the number of points that fall into a circle.  We should be able to get $\pi$ from this.

```python
import numpy
```

```python
N = 500
```

```python
hits = 0
points = [ ]
```

```python
for i in range(N):
    x = numpy.random.random()  # random numbers from zero to one
    y = numpy.random.random()
    if x**2 + y**2 < 1:
        points.append((x, y, True))
        hits += 1
    else:
        points.append((x, y, False))
```

```python
hits
```

```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot
```

```python
x, y, colors = zip(*points)
pyplot.scatter(x, y, c=colors)
```

## $\pi$ formula
* area of square = 1
* area of circle = $\pi r^2 / 4$
* area of circle / area of square = hits / N
* ($\pi r ^2 / 4) / 1 = hits/N$
* $\pi = hits/N * 4$

```python
4 * hits / N 
```
